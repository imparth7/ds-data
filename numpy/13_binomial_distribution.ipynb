# Binomial Distribution

discrete dist - binary output.

param - n(number of trials), p(probability), size(shape-returned array)

## Difference between binomial and normal distribution

Normal - Continous and Binomial - Discrete

use 500 data point for binomial and under 100 data point for normal dist.

given 10 trial for a coin which will generate 10 data points:

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

Given 10 trial for a coin which will generate 10 data point:

```python
x = random.binomial(n=10, p=0.5, size=10)
x
```

Visualization of binomial dist

```python
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)
plt.show()
```

## Difference between binomial and normal distribution
Normal - Continous and Binomial - Discrete

use 500 data point for binomial and under 100 data point for normal dist.

```python
sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='Normal')
plt.show()
```

```python
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='Normal')
plt.show()
```

```python
sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='Normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='Binomial')
plt.show()
```
