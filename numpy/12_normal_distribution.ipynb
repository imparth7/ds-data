# Normal(Gaussian) Distribution

`random.nomral()` method - `loc(mean)`, `scale(standard deviation)`, size

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

We are generating a random nomral distribution of size 2*3.

```python
x = random.normal(size=(2, 3))
x
```

We will generate a random normal dist of size 2*3 with mean at 1 and standard deviation of 2.

```python
x = random.normal(loc=1, scale=2, size=(2, 3))
x
```

### Visualization of normal distribution

```python
sns.distplot(random.normal(size=1000), hist=False)
plt.show()
```

The Curve of the normal dist is also called as bell curve.
